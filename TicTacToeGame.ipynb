```python
from IPython.display import clear_output
import random
board = ['1','2','3','4','5','6','7','8','9']
board_space = [False]*9
```

```python
def display_board(board):
    print(' ___________ ')
    print('|   |   |   |')
    print(f'| {board[6]} | {board[7]} | {board[8]} |')
    print('|___|___|___|')
    print('|   |   |   |')
    print(f'| {board[3]} | {board[4]} | {board[5]} |')
    print('|___|___|___|')
    print('|   |   |   |')
    print(f'| {board[0]} | {board[1]} | {board[2]} |')
    print('|___|___|___|')
```

```python
display_board(board)
```

```python
def player_input(marker):
    position = ' '
    while position not in board:
        position = input(f'Enter your position ({marker}): ')
        if position.isdigit() == False:
            clear_output()
            print('Sorry, your entry is not is a valid digit (1-9)')             
    return int(position)
```

```python
def player_marker(player):
    marker = input(f'Player {player}, enter your marker: ')
    return marker
```

```python
def win_check(board):
    if (board[0]==board[1]==board[2]) or (board[3]==board[4]==board[5]) or (board[6]==board[7]==board[8]) or (board[0]==board[3]==board[6]) or (board[1]==board[4]==board[7]) or (board[2]==board[5]==board[8]) or (board[0]==board[4]==board[8]) or (board[2]==board[4]==board[6]):
        return True
    else:
        return False
```

```python
def full_board_check(board_space):
    if all(board_space):
        return True
    else:
        return False
```

```python
def choose_first():
    return random.randint(1, 2)
```

```python
def replay():
    play = ' '
    while play!='Y' and play!='N':
        play = input('Do you want to play? (Y or N) ').upper()
        if play not in ['Y','N']:
            clear_output()
            print('Sorry, you did not enter a value response')
        if play=='Y':
            return True
        elif play=='N':
            return False
```

```python
def new_game():
    board = ['1','2','3','4','5','6','7','8','9']
    board_space = [False]*9
    clear_output()

    print('Welcome to Tic Tac Toe!')
    playerFirst = choose_first()
    if playerFirst==1:
        playerSec=2
    else:
        playerSec=1

    print(' ')
    print(f'Player {playerFirst} goes is goes first')
    print(f'Player {playerSec} goes is goes second')

    playerFirst_marker = player_marker(playerFirst)
    playerSec_marker = player_marker(playerSec)
    return board, board_space, playerFirst, playerSec, playerFirst_marker, playerSec_marker
```

**Main Script:**

```python
clear_output()

game_on = True

board, board_space, playerFirst, playerSec, playerFirst_marker, playerSec_marker = new_game()



while game_on:
    clear_output()
    display_board(board)

    print(' ')
    print(f'Player {playerFirst} Turn:')
    position = player_input(playerFirst_marker)
    board[position-1]=playerFirst_marker
    board_space[position-1]=True
    display_board(board)    

    if win_check(board):
        print(' ')
        print('-'*21)
        print(f'|  Player {playerFirst}: WINS!  |')
        print('-'*21)
        print(' ')
        game_on = replay()
        if game_on==True:
            board, board_space, playerFirst, playerSec, playerFirst_marker, playerSec_marker = new_game()
            clear_output()
            continue
        else:
            break

    if full_board_check(board_space):
        print(' ')
        print('-'*21)
        print(f'|     Game Over     |')
        print('-'*21)
        print(' ')
        game_on = replay()
        if game_on==True:
            board, board_space, playerFirst, playerSec, playerFirst_marker, playerSec_marker = new_game()
            clear_output()
            continue
        else:
            break
    
    clear_output()
    display_board(board)
    
    print(' ')
    print(f'Player {playerSec} Turn:')
    position = player_input(playerSec_marker)
    board[position-1]=playerSec_marker
    board_space[position-1]=True
    display_board(board)    
    
    if win_check(board):
        print(' ')
        print('-'*21)
        print(f'|  Player {playerSec}: WINS!  |')
        print('-'*21)
        print(' ')
        game_on = replay()
        if game_on==True:
            board, board_space, playerFirst, playerSec, playerFirst_marker, playerSec_marker = new_game()
            clear_output()
            continue
        else:
            break

    if full_board_check(board_space):
        print(' ')
        print('-'*21)
        print(f'|     Game Over     |')
        print('-'*21)
        print(' ')
        game_on = replay()
        if game_on==True:
            board, board_space, playerFirst, playerSec, playerFirst_marker, playerSec_marker = new_game()
            clear_output()
            continue
        else:
            break
```
